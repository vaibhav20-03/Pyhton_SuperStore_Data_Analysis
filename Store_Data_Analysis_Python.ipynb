## Changing required Datatype

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

## Drop supporting columns

```python
df.drop(columns=['Row ID','Ship Mode','Order Priority','Shipping Cost'],inplace=True)
```

## Exploratory Data Analysis---------

### What is the total number of sales?

```python
df['Sales'].sum()
```

### What is the average sales per month? 

```python
df['year']=df['Order Date'].dt.year
df['Month']=df['Order Date'].dt.month
```

```python
df.groupby(['year','Month'])['Sales'].mean()
```

### What is the monthly revenue?

```python
df['Revenue']=df['Sales']*df['Quantity']
```

```python
df.groupby(['year','Month'])['Revenue'].sum()
```

### Which market generated the most sales?

```python
sales_max=df.groupby('Market')['Sales'].sum()
```

```python
sales_max
```

```python
print("Market with highest sales {}  with value of {}".format(sales_max.idxmax(),sales_max.max()))
```

### What were the profits by segment?

```python
segment_df=df.groupby("Segment")['Profit'].sum()
segment_df
```

```python
segment_df.plot(kind='bar',color="orange")
plt.ylabel("Profit")
plt.title("Profits across different Segments")
plt.show()
```

### Which products sell best? 

```python
product_max=df.groupby("Product Name")['Sales'].sum().sort_values(ascending=False)
```

```python
product_max
```

```python
print("Product with highest sales {}  with value of {}".format(product_max.idxmax(),product_max.max()))
```

### What product sold the most? 

```python
sold_max=df.groupby("Product Name")['Quantity'].sum().sort_values(ascending=False)
```

```python
sold_max
```

```python
print("Product with highest Quantity ordered {}  with value of {}".format(sold_max.idxmax(),sold_max.max()))
```

```python
df.head()
```

### Which product categories contribute the most to overall sales?

```python
cate_sum=df.groupby('Category')['Sales'].sum().reset_index()
```

```python
cate_sum
```

```python
df['Sales'].sum()
```

```python
def category_sales(sales):
    return ((sales/df['Sales'].sum())*100).round(2)
        
cate_sum['percentage']=cate_sum['Sales'].apply(lambda x : category_sales(x))     
```

```python
cate_sum
```

```python
plt.pie(cate_sum['percentage'],labels = cate_sum['Category'])
plt.show()
```

### What is the distribution of sales across different regions.

```python
region_df=df.groupby('Region')['Sales'].sum()
region_df
```

```python
region_df.plot(kind='bar')
plt.ylabel("Sales")
plt.title("Sales across different Regions")
plt.show()
```

### What were the sales in different Years?

```python
yearly_sales=df.groupby('year')['Sales'].sum().round(1).reset_index()
```

```python
yearly_sales
```

```python
plt.plot(yearly_sales['year'],yearly_sales['Sales'],marker='o',linestyle='-',color='red')
plt.xlabel("Years")
plt.ylabel("Sales (in Lakhs)")
plt.title("Sales in different Years")
plt.show()
```

### What were the profits in different Years?

```python
yearly_profit=df.groupby('year')['Profit'].sum().reset_index()
```

```python
yearly_profit
```

```python
plt.plot(yearly_profit['year'],yearly_profit['Profit'],marker='o',linestyle='-',color='blue')
plt.xlabel("Years")
plt.ylabel("Profits (in Lakhs)")
plt.title("Profits in different Years")
plt.show()
```

### Top 5 countries with highest sales.

```python
top5_sales=df.groupby("Country")['Sales'].sum().round(2).reset_index().sort_values(by='Sales',ascending=False).head()
```

```python
top5_sales
```

### How is the product price varying with sales.Is there any increase in sales with the decrease in price.

```python
df['Price Each']=df['Sales']/df['Quantity']
```

```python
df['Price Each']
```

```python
pro_price=df.groupby(['year','Product Name'])['Price Each'].sum().reset_index()
pro_price
```

```python
pro_sales=df.groupby(['year','Product Name'])['Sales'].sum().reset_index()
pro_sales
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pro_price['Price Each'],pro_sales['Sales'],color='g')
plt.title("Relation between Price Each and Sales of a Product")
plt.xlabel("Prices")
plt.ylabel("Sales")
plt.show()
```

### What is the average delivery time across the countries and show top 10 countries with lowest delivery time.

```python
df['Delivery time']=df['Ship Date']-df['Order Date']
df['Delivery time']
```

```python
df.groupby("Country")['Delivery time'].mean()
```

```python
country_delivery=df.groupby("Country")['Delivery time'].sum().sort_values().reset_index().head(10)
country_delivery
```

```python
plt.figure(figsize=(10,6))
plt.bar(country_delivery['Country'],country_delivery['Delivery time'],color='m')
plt.title("Delivery time across countries (Top 10)")
plt.xlabel("Countries")
plt.ylabel("Delivery Time (in Days)")
plt.xticks(rotation='vertical')
plt.show()
```
